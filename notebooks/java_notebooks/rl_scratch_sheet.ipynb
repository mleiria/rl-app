```java
import ai.djl.ndarray.NDArray;
import ai.djl.ndarray.NDManager;
import ai.djl.ndarray.types.Shape;

NDManager manager = NDManager.newBaseManager();
// Flatten the 3D array into a 1D array ---
int size = 3 * 3 * 3;
double[] flattenedRewards = new double[size];
int index = 0;
    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            for (int k = 0; k < 3; k++) {
                flattenedRewards[index++] = rewards[i][j][k];
            }
        }
    }

NDArray transitionProbabilitiesND = manager.create(flattenedRewards).reshape(3, 3, 3);
System.out.println("--- Final 3D NDArray ---");
System.out.println(transitionProbabilitiesND);

Shape shape = transitionProbabilitiesND.getShape();
System.out.println("\n--- Shape of the NDArray ---");
System.out.println(shape); // Expected: (3, 3, 3)

System.out.println("\n--- Verifying a value ---");
System.out.println("Value at (1, 2, 0): " + transitionProbabilitiesND.get("1, 2, 0").getDouble()); // Expected: 15.0

/* 
NDArray rewardsND = manager.create(rewards);
NDArray possibleActionsD = manager.create(possibleActions);
*/
```
